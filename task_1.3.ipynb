```python
# task 1.3
# data provided for the input
requestSpending = {
    "Mahek": {
        "balance": 3000.00,
        "transactions": [
            {"amount": -9000.00, "category": "Creatives"},
            {"amount": 7000.00, "category": "Sponsor"},
            {"amount": -2000.00, "category": "Prize-Money"}
        ]
    },
    "Arham": {
        "balance": 5000.00,
        "transactions": [
            {"amount": 8000.00, "category": "Stalls"},
            {"amount": 7500.00, "category": "Seminars"}
        ]
    },
    "Unnati": {
        "balance": 3500.00,
        "transactions": [
            {"amount": -5000.00, "category": "Attraction"},
            {"amount": 2500.00, "category": "Promo"},
            {"amount": -900.00, "category": "Lighting"},
            {"amount": -3000.00, "category": "Games"}
        ]
    },
    "Gaurang": {
        "balance": 2000.00,
        "transactions": [
            {"amount": -1500.00, "category": "Website"},
            {"amount": -1000.00, "category": "C2C"},
            {"amount": -500.00, "category": "Posters"}
        ]
    }
}

""" calcalateTotalSpending wale function mai it calculates the total amount spent of a specific account for the given category,
iterates through the specified account and badmai overall -ve wale(spent ones) ko add up kar dega for the given category"""

def calculateTotalSpending(requestSpending, accountId, category):
    totalSpending = 0
    if accountId in requestSpending:
        transactions = requestSpending[accountId]["transactions"]
        for transaction in transactions:
            if transaction["category"] == category and transaction["amount"] < 0:
                totalSpending += transaction["amount"]
    return totalSpending

""" calculateAccountBalance function does the task of calculating the final balance of specifc account after his/her all transactions are done,
starting with initial balance and adds each transaction amt."""

def calculateAccountBalance(requestSpending, accountId):
    if accountId in requestSpending:
        initialBalance = requestSpending[accountId]["balance"]
        transactions = requestSpending[accountId]["transactions"]
        finalBalance = initialBalance  # yeh isly kiya coz initial balance bhi final balance mai aajaye fir total account balance karenge
        for transaction in transactions:
            finalBalance += transaction["amount"]
        return finalBalance
    return 0

""" calculateMoneyOwed function works for how much is owed for a specific account,
it basically uses the calculateAmountBalance function ,since it has final balance so it will return the abs value if its negative ,which indicates the money is owed"""

def calculateMoneyOwed(requestSpending, accountId):
    finalBalance = calculateAccountBalance(requestSpending, accountId)
    if finalBalance < 0:
        return abs(finalBalance)
    return 0

#taking input for the given data
    
accountId = input("Enter the account ID to calculate spending and balance (Mahek, Arham, Unnati, Gaurang): ")
category = input("Enter the category for calculating the total spending of it: ")

print(f"Total spending in category '{category}': {calculateTotalSpending(requestSpending, accountId, category)}")
print(f"Final balance for account '{accountId}': {calculateAccountBalance(requestSpending, accountId)}")
print(f"Money owed for account '{accountId}': {calculateMoneyOwed(requestSpending, accountId)}")
```
